```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

```python
df = sns.load_dataset('iris')  # Replace 'iris' with the name of the dataset you want to load
print(df.tail())  # Display the first few rows of the dataset
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.head())
```

```python
# Bar plot
sns.barplot(x='species', y='sepal_length', data=df,  palette='turbo')

plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# Calculate the mean of sepal_length and sepal_width for each species
species_means = df.groupby('species')[['sepal_length', 'sepal_width']].mean().reset_index()

# Line plot
sns.lineplot(x='species', y='sepal_length', data=species_means, marker='o', label='Sepal Length')
sns.lineplot(x='species', y='sepal_width', data=species_means, marker='s', label='Sepal Width')

plt.title('Mean Sepal Length and Width by Species')
plt.xlabel('Species')
plt.ylabel('Measurement')
plt.legend()
plt.show()
```
